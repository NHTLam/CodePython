```python
# python artihmetic operations: +, -, *, /, //, %, **
a = int(input('Enter a: '))
b = int(input('Enter b: '))
```

```python
print(f'{a:2} + {b:2} = {a + b:3}')
print('{:2} - {:2} = {:3}'.format(a, b, a-b))
print(f'{a:2} * {b:2} = {a * b:3}')
print(f'{a:2} / {b:2} = {a / b:3}')
print(f'{a:2} // {b:1} = {a // b:3}')
print(f'{a:2} % {b:2} = {a % b:3}')
print(f'{a:2} ** {b:1} = {a ** b:3}')
```

```python
math = float(input('Enter math grade: '))
english = float(input('Enter english grade: '))
science = float(input('Enter science grade: '))
print(f'Math grade: {math}, English grade: {english}, Science grade: {science}')
avg = (math + english + science)/3
print(f'Average = {avg:.2f}')
if avg >= 8.0:
    print('Distinction')
elif avg >= 6.5:
    print('Merit')
elif avg >= 4.0:
    print('Pass')
else:
    print('Fail')
```

```python
print('Distinction' if avg >= 8.0 else 'Merit' if avg >= 6.5 else 'Pass' if avg >= 4.0 else 'Fail')
```

```python
#input:
movieName = input('Movie Name: ')
adultPrice = float(input('Adult Price: '))
childPrice = float(input('Child Price: '))
numberOfAdultTicketSold = int(input('Number of adult ticket sold: '))
numberOfChildTicketSold = int(input('Number of child ticket sold: '))
percentage = float(input('Percentage of Gross Amount Donated: '))
```

```python
total = '$' + str(adultPrice*numberOfAdultTicketSold + childPrice*numberOfChildTicketSold)
amountDonated = '$' + str((adultPrice*numberOfAdultTicketSold + childPrice*numberOfChildTicketSold) * 0.1)
per = str(percentage) + '%'
netSale = '$' + str((adultPrice*numberOfAdultTicketSold + childPrice*numberOfChildTicketSold) - (adultPrice*numberOfAdultTicketSold + childPrice*numberOfChildTicketSold)*0.1)
print ('-*' * 30)
print(f'Movie Name:........................{movieName:>25}')
print(f'Number of Tickets Sold:............{numberOfAdultTicketSold + numberOfChildTicketSold:>25}')
print(f'Gross Amount:......................{total:>25}')
print(f'Percentage of Gross Amount Donated:{per:>25}')
print(f'Amount Donated:....................{amountDonated:>25}')
print(f'Net Sale:..........................{netSale:>25}')
```
