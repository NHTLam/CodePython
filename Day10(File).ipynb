```python
student_data = np.loadtxt('gradebook.csv', delimiter=',', skiprows=1,dtype=str)
#Clean std data
for student in student_data:
    if student[4] == '':
        student[4] = '0'
    if float(student[4]) > 10:
        student[4] = '10'
    if student[2] == '':
        student[2] = 'GCH22000'

#Print std data
# for student in student_data:
#     if float(student[4]) >= 6.5:
#         print(f'{student[2]:10}{student[3]:20}{student[4]:10}')
pass_students = [['Login', 'Email', 'Code', 'Name', 'Grade']]
pass_students += [student for student in student_data if float(student[4]) >=6.5]
for st in pass_students:
    print(f'{st[2]:10}{st[3]:20}{st[4]:10}')

#Write to file 
np.savetxt('pass_student.csv', pass_students, delimiter=',', fmt='%s')
```

```python
codes, names, grades = np.loadtxt('gradebook.csv', delimiter=',', skiprows=1,
                                  usecols=(2,3,4), dtype=str, unpack=True)

for i in range(len(codes)):
    if grades[i] == '':
        grades[i] = '0'
    if codes[i] == '':
        codes[i] = 'GCH22000'
    if float(grades[i]) > 10:
        grades[i] = '10'

# for i in range(len(codes)):
#     if float(grades[i]) >= 6.5:
#         print(f'{codes[i]:10}{names[i]:20}{grades[i]:10}')

# Save clean data into new file
np.savetxt('clean_gradebook.csv', np.column_stack((codes, names, grades)), delimiter=',', fmt='%s', header = 'Code, Name, Grade')
print('Done!')
```

```python
import csv

# Tốn ít bộ nhớ hơn cái trên, nhưng chạy chậm hơn.
with open('gradebook.csv', 'r') as f:   #tranh viec quen close
    reader = csv.reader(f) # -> Tương tự như cái generator
    fw = open('pass_std.csv', 'w')
    writer = csv.writer(fw)

    writer.writerow(next(reader)) # write header

    # next(reader) #skip header
    for row in reader:
        if row[4] == '':
            row[4] = '0'
        if row[2] == '':
            row[2] = 'GCH22000'
        if float(row[4]) > 10:
            row[4] = '10'
        if float(row[4]) >= 6.5:
            # print(f'{row[2]:10}{rows[3]:20}{rows[4]:10}')
            writer.writerow(row)
    fw.close()
print('Done!')
    
```

```python
# Clean pizza_data.csv
f = open('pizza_data.csv', 'r')
reader = csv.reader(f)

# Remove id header, url header
header = next(reader)
del header[0]
del header[7]

fw = open('clean_pizza_data.csv', 'w', encoding='utf-8')
writer = csv.writer(fw)
writer.writerow(header)

for row in reader: 
    # Remove rows that have empty menus.amountMax, menus.amountMin
    if row[9] == '' or row[10] == '':
        continue

    # Rows that have empty priceRangeMax, priceRangeMin ->set to 0
    row[18] = '0' if row[18] == '' else row[18]
    row[19] = '0' if row[19] == '' else row[19]

    # For row that has more than one categories, keep just the 1st one
    row[2] = row[2].split(',')[0]

    # Remove id column, url column
    del row[0]
    del row[7]

    writer.writerow(row)

print('Done!')
f.close()
fw.close()
```
