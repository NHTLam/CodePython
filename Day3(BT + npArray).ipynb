```python
products = ['pencil', 'eraser', 'ruler', 'pen', 'notebook']
prices = [1000, 500, 2000, 5000, 10000]
sold = [120, 1000, 80, 50, 30]
# 120 000, 500 000, 160 000, 250 000, 300 000

# Find product that has the highest sold
max_sold = prices[0] * sold[0]
i_max = 0
n = len(products)
for i in range(1, n):
    if prices[i] * sold[i] > max_sold:
        max_sold = prices[i] * sold[i]
        i_max = i
    
print(f'{products[i_max]} has the highest sold: {max_sold}')

# Find product that has the lowest sold
lowest_sold = prices[0] * sold[0]
i_lowest = 0
for i in range(1, n):
    if prices[i] * sold[i] < lowest_sold:
        lowest_sold = prices[i] * sold[i]
        i_lowest = i
    
print(f'{products[i_lowest]} has the lowest sold: {lowest_sold}')

# Find product that has the highest price
# Find product that has the lowest price
i_highest_price = 0
i_lowest_price = 0
highest_price = prices[0]
lowest_price = prices[0]
for i in range(1, n):
    if prices[i] > highest_price:
        highest_price = prices[i]
        i_highest_price = i
    else:
        lowest_price = prices[i]
        i_lowest_price = i

print(f'{products[i_highest_price]} has the highest price: {highest_price}')
print(f'{products[i_lowest_price]} has the lowest price: {lowest_price}')

# Find product that has sold the most
# Find product that has sold the least
i_sm = 0
i_sl = 0
most_sold = sold[0]
least_sold = sold[0]
for i in range(1, n):
    if sold[i] > most_sold:
        most_sold = sold[i]
        i_sm = i
    else:
        least_sold = sold[i]
        i_sl = i

print(f'{products[i_sm]} has sold the most: {most_sold}')
print(f'{products[i_sl]} has sold the least: {least_sold}')
```
