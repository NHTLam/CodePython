```python
rooms = {}
# rooms = {'Emma': 308, 'John': 230, 'Marry': 111}
rooms = {308:'Emma', 230: 'John', 111: 'Marry'}
print(rooms)
print(len(rooms))
# print(rooms[0]) => error, dictionary doesn't support indexing
print(rooms[308]) # => 308
# print(rooms['emma']) => error because emma doesn't exist, key is case sensitive
```

```python
# Dictionary operations: add new key-value pair, edit value, delete key-value pair
rooms['Mike'] = 123
print(rooms)
```

```python
rooms['Mike'] = 321
print(rooms)
```

```python
del rooms['Mike']
print(rooms)
```

```python
# A new guest arrive
name = input('Enter guest name: ')
room_no = int(input('Enter room number: '))

def add_room(name, room_no):
    if name in rooms:
        print(f'Sorry, guest {name} is already booked')
    else:
        rooms[name] = room_no
        print(f'Guest {name} is booked in room {room_no}')
        
add_room(name, room_no)
```

```python
print(rooms)
```

```python
# A guest wants to change to a new room
name = input('Enter guest name: ')
if name in rooms:
    room_no = int(input('Enter new room number: '))
    add_room(name, room_no)
else:
    print(f'Sorry, guest {name} is a new guest')

print(rooms)
```

```python
# A guest wants to check out
name = input('Enter guest name: ')
if name in rooms:
    del rooms[name]
else:
    print(f'Sorry, guest {name} is a new guest, and did not book a room')

print(rooms)
```
