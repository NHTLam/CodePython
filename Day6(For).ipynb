```python
#General syntax: for <variable> in <iterable>:
for i in range(10):      # ~ for (int i=0; i< 10; i++)
    print(i, end= ' ')
```

```python
a = [1, 2, 3, 4, 5]
for i in range(len(a)):     # ~ for (int i=0; i< a.length; i++)
    print(f'a[{i}] = {a[i]}')
```

```python
for n in a:         # foreach (int n in a)
    print(n)
```

```python
s = 'hello world'
for i, c in enumerate(s): 
    print(f'{i}: {c}')
```

```python
rooms = {'Emma': 308, 'John': 230, 'Marry': 111}
for guest in rooms:
    print(f'{guest} is in room {rooms[guest]}')
print('---------------')
for guest, room_no in rooms.items():
    print(f'{guest} is in room {room_no}')
```

```python
stds = {'John': 8.3, 'Linh': 9.3, 'Phong': 3.4, 'Hoa': 4.6, 'Hanh': 7.8, 'Trung': 5.0, 'Phi': 6.4 }
n_D = 0
n_M = 0
n_P = 0
n_F = 0
for name, gpa in stds.items():
    if gpa >= 8.0:
        n_D += 1
        print(f'Distionction students: {name}')
    elif gpa > 6.5 and gpa < 8.0:
        n_M += 1
        print(f'Merit students: {name}')
    elif gpa >= 4.0 and gpa <= 6.5:
        n_P += 1
        print(f'Pass students: {name}')
    else:
        n_F += 1
        print(f'Fail student: {name}')
print('------------------------------')
print(f'Number of D students: {n_D}')
print(f'Number of M students: {n_M}')
print(f'Number of P students: {n_P}')
print(f'Number of F students: {n_F}')
```

```python
# List comprehension: Generate 1 list based on another list 
# Syntax: [x for y in z]
names = ['John Lennon', 'Paul McCartney', 'George Harrison', 'Ringo Starr']
initials = [n[0] for n in names]
print(initials)
lengths = [len(n) for n in names]
print(lengths)
```

```python
import random as rd
numbers = [rd.randint(-100, 100) for _ in range(10)]
print(numbers)
positives = [n for n in numbers if n > 0]
print(positives)
negative = [n for n in numbers if n < 0]
print(negative)
```

```python
print(numbers)
# convert items of a list 
new_number = [n if n > 0 else 0 for n in numbers if n % 2 == 0]
print(new_number)
```

```python
numbers = [rd.randint(-100, 100) for i in range(20)]
print(numbers)
a = [n for n in numbers if n > 50]
print(a)

mean_list = sum(numbers) / len(numbers)
print(mean_list)
b = [n for n in numbers if n > mean_list]
print(b)

c = [n for n in b if n > 0]
print(c)

d1 = [n for n in numbers if n % 2 == 0]
d2 = [n for n in numbers if n % 2 != 0]
d = d1 + d2
print(d)

even_index = numbers[::2]
print(even_index)
even_index = [numbers[i] for i in range(0, len(numbers), 2)]
print(even_index)
```
