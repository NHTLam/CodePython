```python
def swap(a, b):
    t = a
    a = b
    b = t
    print(f'Inside swap: a = {a}, b = {b}')
```

```python
a, b = 10, 20
swap(a, b)
print(f'Outside swap: a = {a}, b = {b}')
```

```python
def swap_list(a, b):
    n = len(a)
    for i in range(n):
        a[i], b[i] = b[i], a[i]
    print(f'Inside swap_list: a = {a}, b = {b}')
```

```python
a = [1 ,2 ,3]
b = [4, 5 ,6]
swap_list(a, b)
print(f'Outside swap_list: a = {a}, b = {b}')
```

```python
def min_max(*items):
    mi = ma = items[0]
    for i in items:
        if i < mi:
            mi = i
        if i > ma:
            ma = i 
    return mi, ma 
```

```python
# call min_max, store the result in 2 variables
a, b = min_max(3, 4, 5, 1, -3, 0)
print(f'Min = {a}, Max = {b}')
# call min_max, store the result in a single variable
a, _= min_max(3, 4, 5, 1, -3, 0)
print(f'Min = {a}')
_, b = min_max(3, 4, 5, 1, -3, 0)
print(f'Max = {b}')
c = min_max(3, 4, 5, 1, -3, 0)
print(c)
```

```python
def greater(x, y): 
    return x > y

def less(x, y):
    return x < y

def greater_last_digit(x, y):
    return x % 10 > y % 10

def my_sort(items, compare = greater):
    for i in range(len(items) - 1):
        for j in range(i + 1, len(items)):
            if compare(items[i], items[j]):
                items[i], items[j] = items[j], items[i]
```

```python
import random as rd 
numbers = [rd.randint(1, 100) for _ in range(10)]
print(numbers)
my_sort(numbers)
print(numbers)
my_sort(numbers, less)
print(numbers)
my_sort(numbers, greater_last_digit)
print(numbers)
```

```python
def greater_len(x,y):
    return len(x) > len(y)
names = ['John Murray', 'Adam Smith', 'Marrie Jane', 'Harry Miller']
my_sort(names)
print(names)
my_sort(names, greater_len)
print(names)
```

```python
# Sort names by number of vowels
def numbers_of_vowels(x):
    count = 0
    for i in x:
        if i in 'ueoai':
            count += 1
    return count

def greater_number_of_vowels(x,y):
    a = numbers_of_vowels(x)
    b = numbers_of_vowels(y)
    return a > b

# Bestter way: 
def greater_number_of_vowels_Best(x,y):
    return len([c for c in x if c in 'aeiou']) > len([c for c in y if c in 'aeiou'])
my_sort(names, greater_number_of_vowels_Best)
print(names)

my_sort(names, greater_number_of_vowels)
print(names)

# Sort names by last name
def greater_lastname(x, y):
    return x.split()[-1] > y.split()[-1]
my_sort(names, greater_lastname)
print(names)
```
