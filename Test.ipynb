```python
mylist = ["Blue", "Red", "Green"]

# Chèn chuỗi 'White' vào trước phần tử có index bằng 2
mylist.insert(2, "White")
print(mylist)
#>> ['Blue', 'Red', 'White', 'Green']

# Chèn chuỗi 'Black' vào đầu list
mylist.insert(0, "Black")
print(mylist)
#>> ['Black', 'Blue', 'Red', 'White', 'Green']
```

```python
lis = [1, 2, 3, 4, ' ']
t = lis[0]
for i in range(len(lis) -1):
    o = lis[i+1]
    lis[i+1] = t
    t = o
print(lis)
```
